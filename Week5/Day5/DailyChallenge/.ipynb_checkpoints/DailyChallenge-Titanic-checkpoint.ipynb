# 5

### survival rate by genders

```python
surv_sex = df.groupby(['Survived', 'Sex']).count().reset_index()
surv_sex
```

```python
fem_surv = surv_sex.loc[(surv_sex['Survived'] == 1) & (surv_sex['Sex'] == 'female')]['PassengerId'].iloc[0]
mal_surv = surv_sex.loc[(surv_sex['Survived'] == 1) & (surv_sex['Sex'] == 'male')]['PassengerId'].iloc[0]
```

```python
fem_amount = fem_surv + surv_sex.loc[(surv_sex['Survived'] == 0) & (surv_sex['Sex'] == 'female')]['PassengerId'].iloc[0]
mal_amount = mal_surv + surv_sex.loc[(surv_sex['Survived'] == 0) & (surv_sex['Sex'] == 'male')]['PassengerId'].iloc[0]
```

```python
# FEMALE SURVIVING RATE
fem_serv_rate = fem_surv / fem_amount
fem_serv_rate
```

```python
# MALE SURVIVING RATE ;(
mal_serv_rate = mal_surv / mal_amount
mal_serv_rate
```

### survival rate by passenger classes

```python
surv_pc = df.groupby(['Survived', 'Pclass']).count().reset_index()
surv_pc
```

```python
firstpc_surv = surv_pc.loc[(surv_pc['Survived'] == 1) & (surv_pc['Pclass'] == 1)]['PassengerId'].iloc[0]
secondpc_surv = surv_pc.loc[(surv_pc['Survived'] == 1) & (surv_pc['Pclass'] == 2)]['PassengerId'].iloc[0]
thirdpc_surv = surv_pc.loc[(surv_pc['Survived'] == 1) & (surv_pc['Pclass'] == 3)]['PassengerId'].iloc[0]


firstpc_amount = firstpc_surv + surv_pc.loc[(surv_pc['Survived'] == 0) & (surv_pc['Pclass'] == 1)]['PassengerId'].iloc[0]
secondpc_amount = secondpc_surv + surv_pc.loc[(surv_pc['Survived'] == 0) & (surv_pc['Pclass'] == 2)]['PassengerId'].iloc[0]
thirdpc_amount = thirdpc_surv + surv_pc.loc[(surv_pc['Survived'] == 0) & (surv_pc['Pclass'] == 3)]['PassengerId'].iloc[0]

firstpc_serv_rate = firstpc_surv / firstpc_amount
secondpc_serv_rate = secondpc_surv / secondpc_amount
thirdpc_serv_rate = thirdpc_surv / thirdpc_amount
```

```python
# SURVIVAL RATES 1-2-3 Passanger classes
print(firstpc_serv_rate, secondpc_serv_rate, thirdpc_serv_rate)
```
