### 4

```python
def filter(x):
    if x <= 50:
        return 'new'
    if (x > 50 and x <= 70):
        return 'old'
    if x > 70:
        return 'very old'

boston['age_cat'] = boston['age'].apply(filter)
boston['age_cat']
```

```python
# y = pd.get_dummies(boston.age_cat)
# y
boston['age_cat'].replace(['new', 'old', 'very old'], 
                                              [0, 1, 2], inplace=True)
boston['age_cat']
```

# Part 2

### 1

```python
new_properties = boston[boston['age'] <= 35]
old_properties = boston[(boston['age'] > 35) & (boston['age'] <= 70)]
very_old_properties = boston[boston['age'] > 70]
```

```python
figure, axes = plt.subplots()
axes.set(title='New', xlabel='index', ylabel='age')
axes.plot(new_properties['age'])
axes.axhline(y=np.median(new_properties.age), color='red', linestyle='--', linewidth=3, label='Avg')
```

```python
figure, axes = plt.subplots()

axes.set(title='Old', xlabel='index', ylabel='age')
axes.plot(old_properties['age'])
axes.axhline(y=np.median(old_properties.age), color='red', linestyle='--', linewidth=3, label='Avg')
```

```python
figure, axes = plt.subplots()

axes.set(title='Very Old', xlabel='index', ylabel='age')
axes.plot(very_old_properties['age'])
axes.axhline(y=np.median(very_old_properties.age), color='red', linestyle='--', linewidth=3, label='Avg')
```

# Part 3

### 1

```python
figure, axes = plt.subplots()
boston['crim_bin'] = pd.cut(boston['crim'], bins=[0, 1, 5, 10, 20, max(boston['crim'])], 
                        labels=['0-1', '1-5', '5-10', '10-20', '20+'])
boston['crim_bin'].head(20)
plt.hist(boston.crim_bin, bins = 5, edgecolor="k")
```

```python
boston_grouped = boston.groupby('crim_bin')['medv'].mean().reset_index()
boston_grouped
```

```python
figure, axes = plt.subplots()
axes.set(title="crimes to price", xlabel="price", ylabel='crimes')
axes.plot(boston_grouped['medv'], boston_grouped['crim_bin'])
```
