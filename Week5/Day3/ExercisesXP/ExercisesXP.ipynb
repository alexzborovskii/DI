# EX7

```python
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

figure, axes_arr = plt.subplots(1,2)
axes_arr[0].set(title="y = sin(x)")
axes_arr[1].set(title="y = sin(x)")
axes_arr[0].plot(x, y1, color='blue', label = 'line')
axes_arr[1].scatter(x, y2, color='blue', label = 'cos')
```

# EX9

```python
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

figure, axes_arr = plt.subplots(2)
axes_arr[0].plot(x, y1, color='blue',  label='y = sin(x)')
axes_arr[1].scatter(x, y2, color='blue',  label='y = cos(x)')
axes_arr[0].legend(loc='upper left')
axes_arr[1].legend(loc='center right')
plt.tight_layout() 
```
