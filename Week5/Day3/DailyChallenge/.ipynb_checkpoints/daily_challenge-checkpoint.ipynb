```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Task 1

```python
x = np.linspace(0, 20, 100)
x
```

```python
y = x ** 2
y
```

```python
figure, axes = plt.subplots()
axes.set(title='Plot of x vs y (y=x^2)', xlabel='x', ylabel='y', xlim=(0, 20), ylim=(0, 400))
axes.plot(x, y)
figure.savefig('ex2.png')
```

## Task 2

```python
chipo = pd.read_csv('chipotle.txt', sep="\t")
chipo
```

```python
chipo['item_price'] = chipo['item_price'].apply(lambda c: c[1:])
chipo
```

```python
chipo['item_price'] = chipo['item_price'].astype('float32')
```

```python
chipo.info()
```

```python
figure, axes = plt.subplots()
axes.set(title='Relationship between Item Price and Quality Ordered', xlabel='Item Price',
        ylabel='Quantity Ordered')
axes.scatter(chipo['item_price'], chipo['quantity'])

figure.savefig('Task2')
```

## Task 3

```python
chipo_grouped = chipo.groupby('item_name').sum().sort_values('quantity', ascending=False).head(5)
# chipo_grouped.column(['item_name'])
chipo_grouped
```

```python
figure, axes = plt.subplots()
axes.bar(chipo_grouped.index, chipo_grouped['quantity'])
axes.tick_params(axis='x', labelrotation = 45)
axes.set(title='Top 5 Items by Quantity Ordered', xlabel='Item Name', ylabel='Total Quantity Ordered')

plt.tight_layout() # more space between axes
```
