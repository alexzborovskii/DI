## Exercises 7-9

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

figure, axes_set = plt.subplots(2, 2)
axes_set[0][0].plot(y, x, color='green', label = 'line')
axes_set[0][1].plot(-y, y, color='green', label = 'line')
axes_set[1][0].plot(y**2, x, color='green', label = 'line')
axes_set[1][1].plot((-y)**2, y, color='green', label = 'line')
```

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

# figure, axes_set = plt.subplots(2, 2)
figure, axes_set = plt.subplots(2, 2, sharex=True)

# PLOTS

axes_set[0][0].plot(x, y, color='green', label = 'y, x')
axes_set[0][1].plot(x, -y, color='green', label = '-y, x')
axes_set[1][0].plot(x, y**2,  color='green', label = 'y**2, x')
axes_set[1][1].plot(x, -y**2, color='green', label = '-y**2, x')



#MAX, MIN

max_y_idx = np.argmax(y)
min_y_idx = np.argmin(y)

axes_set[0][0].annotate('max', xy=(x[max_y_idx], y[max_y_idx]), xytext=(x[max_y_idx]+1,  y[max_y_idx] + 0.5),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )
axes_set[0][0].annotate('min', xy=(x[min_y_idx], y[min_y_idx]), xytext=(x[min_y_idx]-2,  y[min_y_idx] - 0.5),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )



#LEGENDS

axes_set[0][0].legend(title = 'Legend', loc='lower left', shadow=True)
axes_set[0][1].legend(title = 'Legend', loc='upper right', shadow=True)
axes_set[1][0].legend(title = 'Legend', loc='lower right', shadow=True)
axes_set[1][1].legend(title = 'Legend', loc='upper left', shadow=True)

plt.tight_layout() # more space between axes
```
