```python
x = np.random.rand(100)
y = np.random.rand(100)

figure, axes = plt.subplots()
axes.scatter(x, y)
```

### Histogram - putting the data into bins based on values

```python
# data = np.random.rand(1000)
data = [1, 1, 1, 1, 1, 1, 1, 5, 5, 5, 5, 5, 5, 5, 5, 5, 9, 9, 9, 9, 9]
figure, axes = plt.subplots()
axes.hist(data, bins = 3)
```

## Markers

```python
figure, axes = plt.subplots()
axes.set(title="Line plot", xlabel="X", ylabel="Y")
axes.plot([1, 2, 3, 4],[2, 4, 5, 7])
```

```python
figure, axes = plt.subplots()
axes.set(title="Line plot", xlabel="X", ylabel="Y" )
axes.plot([1, 2, 3, 4],[2, 4, 5, 7], color="green", linestyle="--")
```

```python
figure, axes = plt.subplots()
axes.set(title="Line plot", xlabel="X", ylabel="Y" )
axes.plot([1, 2, 3, 4],[2, 4, 5, 7], color="green", linestyle="--", marker='o', mfc = 'red', mec='blue')
```

```python
figure, axes = plt.subplots()
axes.set(title="Line plot", xlabel="X", ylabel="Y" )
axes.plot([1, 2, 3, 4],[2, 4, 5, 7], color="green", linestyle="--", marker='o', mfc = 'red', mec='blue', label = 'Yossi')
axes.plot([6, 3, 5, 10],[6, 7, 4, 3], color="red", linestyle="--", marker='*', mfc = 'blue', mec='blue', label = 'Ziv')
axes.legend()
plt.show # to show plot in VS Code
```

# Save plot

```python
figure, axes = plt.subplots()
axes.set(title="Line plot", xlabel="X", ylabel="Y" )
axes.plot([1, 2, 3, 4],[2, 4, 5, 7], color="green", linestyle="--", marker='o', mfc = 'red', mec='blue', label = 'Yossi')
axes.plot([6, 3, 5, 10],[6, 7, 4, 3], color="red", linestyle="--", marker='*', mfc = 'blue', mec='blue', label = 'Ziv')
axes.legend()

figure.savefig('my_plot.png')
```

```python
x = np.linspace(0, 10, 20)
y1 = x ** 2
y2 = x ** 1.5
figure, axes = plt.subplots()
axes.set(title="Line plot", xlabel="X", ylabel="Y" )
axes.plot(x, y1, color="green", linestyle="--", marker='o')
axes.scatter(x, y2, color="red", linestyle="--", marker='*')
```

###  EXERCISE 4

```python
x = np.linspace(0, 10, 20)
y1 = x ** 2
y2 = x ** 1.5
figure, axes = plt.subplots()
axes.set(title = "My Styled Plot", xlabel="X", ylabel="Y")
axes.plot(x, y1, color="blue", linestyle="-", marker='o', label = "sqrt")
axes.plot(x, y2, color="red", linestyle=":", marker='s', label = "power 1,5")
axes.legend()
figure.savefig('my_styled_plot.png')
```
